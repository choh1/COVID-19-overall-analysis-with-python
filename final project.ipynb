# Visualization


```python
#A relationship between patients who were tested for postive of COVID-19 and
#who were recovered from COVID-19

sns.relplot(x="positive", y="recovered",data=d1)

#for the initial 1,000,000 number of patients who were had positive test results for COVID-19, 
#there are less than 200,000 recovered. However, as the number grew for positive tests,
#more than 1,200,000 patients has recovered from the disease.
#There is a linear relationship. It can be concluded that people are recovering.
```

```python
#A relationship for the total people who had been tested
#and had positive result for COVID-19.

sns.relplot(x="totalTestResults", y="positive", data=d1)

#Out of the total number of test cases are 4e7, more than 4,000,000 people has been tested for postitive.
```

```python
sns.relplot(x="death", y="positive", 
            data=d1,
            hue="recovered")   #hue shows recovered patients who were tested for postive COVID-19
#It explains that as people are tested with positive, the number of death increases correspondingly.
#However, as the number of death increases, patients being recovered is also increasing. 
```

```python
#relationships between the variables 
sns.pairplot(d1)
```

```python
sns.relplot(x ='Age_group',
            y ='COVID-19_Deaths', 
            kind='line',  #draw a line graph
            data = d2)
```

```python
#categorical data plot
sns.catplot(x='Age_group',
           y='COVID-19_Deaths',
           data=d2)

#The plot indicates that as the age becomes older, the number of deaths increases correspondingly. 
#The number is greatest for age groups of all ages.
```
